Transformations  in the images:


* 		Resizing to 256x256: Resizing all the images. 
* 		Random Affine Transformations: Appliying affine transformations like rotations, translations, and shears, in the range of 10 degrees for rotation and 10% for scale and translation. That kind of transformations help us to increse the robustnesss and we are able to manage variations in the data.
* 		Random Horizontal Flip:  50% probability that a horizontal flip is applied. We can have more variability.
*       RandAugment

Data augmentation with random flips, rotations, and translations, randAugment helps expose the model to diverse examples during training, improving its ability to generalize.
We don´t use this kind of transformations for the validation and test datasets because we want the evaluation with the original.

<img src="static_images/icons/noun-answer-3361020.png" alt=">" style="width:25px"/> __Answer:__ * 

The CCN architecture consists in 5 blocs of Convolutional layers and a fully connected layers. This is how are build.

CONVOLUTIONAL BLOCKS
* 		Input Layer: We have three color channels (RGB):
    * nn.Conv2d(3, 16, 3, padding=1), 3 input channels (colours), 16 output channels, kernel size, 3x3, with a padding=1 and ReLU activations
* 		BatchNormalization: helps doing faster convergence and regularization.
* 		(Max-Pooling): Using a MaxPool layer to reduce the dimensions.
    * 		nn.MaxPool2d(2, 2): Kernel size 2x2 and a stride of 2, we get half dimensions


Doing that 5 times. Each convolutional layer increases the number of channels, getting more features.


FULLY CONNECTED LAYERS:

After the convolutional blocks, there are a fully connected layers to perform the final classification.
Firstly its flattened the feature maps using nn.Flatten().

* 		nn.Linear(256 * 7 * 7, 256): takes the flattened feature maps as input. 
    * 		(7x7 after max-pooling).
* 		 nn.Linear(256, num_classes): The final fully connected layer with the output size equal to num_classes. This is where the classification is performed.
    
* 		 Dropout and Batch Normalization:
     It added dropout layers (nn.Dropout) after the fully connected layers to reduce overfitting. 
     nn.BatchNorm1d: this normalizes the activations and helps in faster convergence and regularization.

The model reduce dimensions. The dropout and normalization are used to improve the model and prevent overfitting


model.py (summary)
  *  Convolutional layers, All the filtered images.
 
  *  BatchNorm,  to normalizes it.
 
  *  Maxpooling layers, which reduce dimensions of an input, keeping only the most active pixels from the previous layer.
 
  *  The usual Linear + Dropout layers to avoid overfitting.
